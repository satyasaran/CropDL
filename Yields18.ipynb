```python
import pandas as pd
import numpy as np
```

```python
yield_data_18=pd.read_csv('./Dataset_Potato_18_20/Yield2018.csv').drop(columns=['Unnamed: 0'])
```

```python
yield_data_18
```

```python
idata=yield_data_18[['bin','bed','row','ID','position','yield','yieldDM']].copy()
```

```python
idata
```

```python
Y#datap_T=idata[(idata['position']=='South')&(idata['bed']==3)| (idata['position']=='North')&(idata['bed']==4)]

```

```python
idata_wet_sample=idata[(idata['bed']==3)&(idata['bin']==4)| (idata['bed']==4)&(idata['bin']==1)]
```

```python
idata_wet_sample
```

```python
idata_dry_sample=idata[(idata['bed']==3)&(idata['bin']==3)| (idata['bed']==4)&(idata['bin']==2)]
```

```python
idata_dry_sample
```

```python
RL=pd.read_csv('SquareRL_2018.csv').drop(columns=['Unnamed: 0'])
```

```python
RL
```

```python
pd.merge(idata_dry_sample,RL,on='row').to_csv('SQRL_Dry_Yields_average_tube_2018.csv')
pd.merge(idata_wet_sample,RL,on='row').to_csv('SQRL_Wet_Yields_average_tube_2018.csv')
```

```python
RL_yields_Dry=pd.merge(idata_dry_sample,RL,on='row')
RL_yields_Wet=pd.merge(idata_wet_sample,RL,on='row')
```

```python
RL_yields_Dry
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
RL_table_main_Wet=RL_yields_Wet.drop(['row','bin','bed','ID','position'], axis=1)
RL_table_main_Dry=RL_yields_Dry.drop(['row','bin','bed','ID','position'], axis=1)

```

```python
RL_table_main_Wet
```

```python
RL_table_main_Wet['TRL']=RL_table_main_Wet.iloc[:,2:].sum(axis=1)
RL_table_main_Dry['TRL']=RL_table_main_Dry.iloc[:,2:].sum(axis=1)
```

```python
RL_table_main_Wet['TRL_June']=RL_table_main_Wet.filter(regex='June').sum(axis=1)
RL_table_main_Wet['TRL_July']=RL_table_main_Wet.filter(regex='July').sum(axis=1)
RL_table_main_Wet['TRL_Aug']=RL_table_main_Wet.filter(regex='Aug').sum(axis=1)

RL_table_main_Dry['TRL_June']=RL_table_main_Dry.filter(regex='June').sum(axis=1)
RL_table_main_Dry['TRL_July']=RL_table_main_Dry.filter(regex='July').sum(axis=1)
RL_table_main_Dry['TRL_Aug']=RL_table_main_Dry.filter(regex='Aug').sum(axis=1)
```

```python
# Removing columns containg zeros
RL_table_main_Dry=RL_table_main_Dry.replace(0,np.nan).dropna(axis=1,how="all")
RL_table_main_Wet=RL_table_main_Wet.replace(0,np.nan).dropna(axis=1,how="all")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(RL_table_main_Wet.corr(), dtype=np.bool))
heatmap = sns.heatmap(RL_table_main_Wet.corr(), mask=mask, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Yields (Wet: bed 3, bin 4; bed 4, bin 1) and Square Root lengths in 2018', fontdict={'fontsize':18}, pad=16);
plt.savefig('Corr_RL_Wet_yields_2018.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(RL_table_main_Dry.corr(), dtype=np.bool))
heatmap = sns.heatmap(RL_table_main_Dry.corr(), mask=mask, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Yields (Dry: bed 3, bin 3; bed 4, bin 2) and Square Root lengths in 2018', fontdict={'fontsize':18}, pad=16);
plt.savefig('Corr_RL_Dry_yields_2018.png')
```
